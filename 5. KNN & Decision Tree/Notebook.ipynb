```python
import pandas as pd
import seaborn as sns
import warnings
from matplotlib import pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split as tts
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, classification_report
from sklearn.metrics import mean_absolute_error, mean_squared_error
warnings.filterwarnings("ignore")
```

# Data Analysis

```python
df = pd.read_excel('cancer.xlsx')
df.head()
```

### Columns

```python
df.columns
```

```python
df.drop('id', axis=1, inplace=True)
df.head()
```

### Check for null

```python
df.isnull().any().values
```

### Categorical columns

```python
df.select_dtypes(include="object").columns
```

```python
df.diagnosis.unique()
```

### Encoding

```python
le = LabelEncoder()
df.diagnosis = le.fit_transform(df.diagnosis)
df.diagnosis.value_counts()
```

```python
df.describe()
```

### Correlation

```python
plt.figure(figsize=(18,12))
plt.title('Correlation for every column')
sns.heatmap(df.corr(), annot=True, cmap="Greens")
```

### Spliting Dataset

```python
X = df.drop('diagnosis', axis=1)
y = df.diagnosis
```

```python
xtrain, xtest, ytrain, ytest = tts(X, y, test_size=0.3, random_state=2)
```

# K-nearest neighbors (KNN)

```python
knn = KNeighborsClassifier()
knn.fit(xtrain,ytrain)
```

```python
knn_pred = knn.predict(xtest)
```

```python
knn.predict([[17.99,10.38,122.8,1001,0.1184,0.2776,0.3001,0.1471,0.2419,0.07871,1.095,0.9053,8.589,153.4,0.006399,0.04904,0.05373,0.01587,0.03003,0.006193,25.38,17.33,184.6,2019,0.1622,0.6656,0.7119,0.2654,0.4601,0.1189]])
```

# Decision tree

```python
dt = DecisionTreeClassifier()
dt.fit(xtrain, ytrain)
```

```python
dt_pred = dt.predict(xtest)
```

```python
dt.predict([[17.99,10.38,122.8,1001,0.1184,0.2776,0.3001,0.1471,0.2419,0.07871,1.095,0.9053,8.589,153.4,0.006399,0.04904,0.05373,0.01587,0.03003,0.006193,25.38,17.33,184.6,2019,0.1622,0.6656,0.7119,0.2654,0.4601,0.1189]])
```

# Evaluate Model

## Classification report

```python
print("\t\t  K-nearest neighbors (KNN)")
print(classification_report(ytest,knn_pred))
```

```python
print("\t\t\tDecision Tree:")
print(classification_report(ytest,dt_pred))
```

## Confusion Matrix

```python
knn_cm = confusion_matrix(ytest, knn_pred)
dt_cm = confusion_matrix(ytest, dt_pred)
```

```python
plt.figure(figsize=(14,4))

#For KNN
plt.subplot(1,2,1)
sns.heatmap(knn_cm, annot=True)
plt.xlabel('Predicted Result')
plt.ylabel('Actual value')
plt.title("KNN Confusion Matrix")

#For Decision Tree
plt.subplot(1,2,2)
sns.heatmap(dt_cm, annot=True)
plt.xlabel('Predicted Result')
plt.ylabel('Actual value')
plt.title("Decision Tree Confusion Matrix")
```

## Accuracy score

```python
print(f"KNN: {accuracy_score(ytest, knn_pred)}")
print(f"Decision Tree: {accuracy_score(ytest, dt_pred)}")
```

## ROC Curve

```python
knn_fpr, knn_tpr, knn_thresholds = roc_curve(ytest, knn_pred)
dt_fpr, dt_tpr, dt_thresholds = roc_curve(ytest, dt_pred)
```

```python
plt.plot(knn_fpr, knn_tpr)
plt.plot(dt_fpr, dt_tpr)
plt.xlabel("False Positive Rate")
plt.ylabel( "True Positive Rate")
plt.title('ROC Curve')
plt.legend(["KNN", "Decision Tree"], loc ="lower right")
```

# Errors

#### KNN

```python
knn_mae = mean_absolute_error(ytest, knn_pred)
knn_mse = mean_squared_error(ytest, knn_pred)
print("Mean absolute error: ", knn_mae)
print("Mean squared error: ", knn_mse)
```

### Decision Tree

```python
dt_mae = mean_absolute_error(ytest, dt_pred)
dt_mse = mean_squared_error(ytest, dt_pred)
print("Mean absolute error: ", dt_mae)
print("Mean squared error: ", dt_mse)
```
